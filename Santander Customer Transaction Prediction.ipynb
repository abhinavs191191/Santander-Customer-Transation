### outlier analysis

```python
col_names = train.drop(['target','ID_code'], axis=1).columns
col_names
```

```python
for i in col_names:
    q75, q25 =np.percentile(train.loc[:,i],[75,25])
    iqr  = q75-q25
    min  = q25 - (iqr*1.5)
    max  = q75 + (iqr*1.5)
    train = train.drop(train[train.loc[:,i]<min].index) 
    train = train.drop(train[train.loc[:,i]>max].index)
```

```python
train.shape
```

```python
for i in col_names:
    q75, q25 =np.percentile(test.loc[:,i],[75,25])
    iqr  = q75-q25
    min  = q25 - (iqr*1.5)
    max  = q75 + (iqr*1.5)
    test = test.drop(test[test.loc[:,i]<min].index) 
    test = test.drop(test[test.loc[:,i]>max].index)
```

```python
test.shape
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = train.drop(['ID_code', 'target'], axis = 1)

y = train['target']
```

```python
logit = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
import warnings as warning
warning.filterwarnings("ignore")
```

```python
logit.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
from sklearn.metrics import confusion_matrix
y_pred = logit.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
print(classification_report(y_test, y_pred))
```

```python
roc_curve(y_test, y_pred)
```

```python
print(roc_auc_score(y_test, y_pred))
```

```python
from flask import Flask, request, jsonify
# Load the model
model = pickle.load(open('model.pkl','rb'))
@app.route('/api',methods=['POST'])
def predict():
    # Get the data from the POST request.
    data = request.get_json(force=True)

    # Make prediction using model loaded from disk as per the data.
    prediction = model.predict([[np.array(data['var_0'])]])

    # Take the first value of prediction
    output = prediction[0]

    return jsonify(output)

if __name__ == '__main__':
    app.run(debug=False)
```
